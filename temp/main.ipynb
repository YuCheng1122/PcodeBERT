```python
import pickle

file = "/home/tommy/Project/PcodeBERT/outputs/preprocessed/pcode_corpus_x86_64_new_data.pkl"

print(f"Inspecting data structure in: {file}")

try:
    with open(file, 'rb') as f:
        # 1. 只載入第一個批次 (batch)
        #    這對應您程式碼中的 'data' 或 'corpus_batch'
        first_batch = pickle.load(f)

        # 2. 檢查這個 'batch' 物件本身的型別
        print(f"Type of the batch object ('data'): {type(first_batch)}")

        # 3. 檢查 'batch' 是否為 list 且不為空
        if isinstance(first_batch, list) and len(first_batch) > 0:
            
            # 4. 取出批次中的第一個元素
            #    這對應您程式碼中的 'sentence'
            first_element = first_batch[0]
            
            # 5. 檢查這個元素的型別 (這是關鍵)
            print(f"Type of the first element in batch ('sentence'): {type(first_element)}")

            # 6. 印出這個元素的部分內容看看
            print(f"Content of the first element (preview):")
            print(first_element)
            
        elif isinstance(first_batch, list) and len(first_batch) == 0:
            print("The first batch is an empty list.")
        else:
            print("The loaded 'data' object is not a list or is empty.")

except EOFError:
    print("Error: The file is empty.")
except Exception as e:
    print(f"An error occurred: {e}")
```
